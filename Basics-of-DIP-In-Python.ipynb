```python
while True:
    cv2.imshow('Image1',img)
    if cv2.waitKey(2) == 27:  #27 Is ASCII value of ESC, c2.waitKey(2), waits for 2 sec between switching windows
        break;
cv2.destroyAllWindows() 
```

```python
haarData = cv2.CascadeClassifier("haarcascade_frontalface_default.xml") #Loading Haar Data
```

```python
haarData.detectMultiScale(img)  #return array[x, y, width, height] of face
```

```python
#cv2.rectangle(img,(x,y),(w,h),(b,g,r),border_thickness)
```

```python
while True:                           #drawing rectangle on face detected
    face = haarData.detectMultiScale(img);
    for x,y,w,h in face:
        cv2.rectangle(img,(x,y),(x+w,y+h),(255,255,0),0)
    cv2.imshow('Image1',img)
    if cv2.waitKey(2) == 27:  #27 Is ASCII value of ESC, c2.waitKey(2), waits for 2 sec between switching windows
        break;
cv2.destroyAllWindows() 
```

```python
#Reading data from camera
cam = cv2.VideoCapture(0)
while True:                           #drawing rectangle on face detected
    flag, img = cam.read()
    if flag:
        face = haarData.detectMultiScale(img);
        for x,y,w,h in face:
            cv2.rectangle(img,(x,y),(x+w,y+h),(255,255,0),0)
        cv2.imshow('Image1',img)
        if cv2.waitKey(2) == 27:  #27 Is ASCII value of ESC, c2.waitKey(2), waits for 2 sec between switching windows
            break;
            
cam.release()
cv2.destroyAllWindows() 
```
